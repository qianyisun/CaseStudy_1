```python
import pandas as pd 
import numpy as np
snf = pd.read_csv("SNFs.csv")
episodes = pd.read_csv("episodes_dur.csv")
```

```python
snf
```

```python
episodes
```

```python
num_patients = episodes.groupby(['PID']).size()
```

```python
num_patients = num_patients.to_frame().reset_index()
num_patients
```

```python
num_patients.columns = ['PID', '#patients']
```

```python
num_patients
```

```python
snf = snf.merge(num_patients, on = 'PID')
```

```python
snf = snf[['PID', 'County', '#patients']]
```

```python
snf
```

```python
pid_per_county = snf.groupby(['County']).size()
```

```python
pid_per_county = pid_per_county.to_frame().reset_index()
pid_per_county.columns = ['County', '#SNFs']
```

```python
patientsbycounty = snf.groupby(['County']).agg('sum').reset_index()
```

```python
patientsbycounty = patientsbycounty.drop(['PID'], axis = 1)
```

```python
county_level_data = pid_per_county.merge(patientsbycounty, on = 'County')
```

```python
county_level_data
```

```python
county_level_data.to_csv('County_Data.csv', index = False)
```

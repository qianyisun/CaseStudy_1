```python
from googlemaps import Client as GoogleMaps
import pandas as pd 
import time
```

```python
gmaps = GoogleMaps('AIzaSyDeEtdjVxFzjs4PuBoF1xVEjfw-qGzqYzo')
```

```python
addresses = pd.read_csv("SNFs.csv")
addresses.head()
```

```python
addresses['Full_Address'] = addresses['ProviderPracticeAddress1'].str.cat(addresses['ProviderPracticeCity'],sep=" ")
addresses.head()
```

```python
addresses['Full_Address'] = addresses['Full_Address'].str.cat(addresses['ProviderPracticeStateCode'],sep=" ")
addresses.head()
```

```python
addresses['ProviderPracticeZipCode'] = addresses['ProviderPracticeZipCode'].astype(str)
```

```python
addresses['Full_Address'] = addresses['Full_Address'].str.cat(addresses['ProviderPracticeZipCode'],sep=" ")
addresses.head()
```

```python
addresses['long'] = ""
addresses['lat'] = ""
```

```python
addresses.head()
```

```python
for x in range(len(addresses)):
    try:
        time.sleep(1) #to add delay in case of large DFs
        geocode_result = gmaps.geocode(addresses['Full_Address'][x])
        addresses['lat'][x] = geocode_result[0]['geometry']['location'] ['lat']
        addresses['long'][x] = geocode_result[0]['geometry']['location']['lng']
    except IndexError:
        print("Address was wrong...")
    except Exception as e:
        print("Unexpected error occurred.", e )
addresses.head()
```

```python
addresses.to_csv('addresses.csv', index = False)
```

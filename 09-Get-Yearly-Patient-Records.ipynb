```python
# load library
import math
import numpy as np
import pandas as pd
```

```python
# load data
data = pd.read_csv('Data/episodes_dur.csv')
```

```python
data.head()
```

```python
data = data[["PID","EpisodeID","ContractType","duration"]]
```

```python
#calculate the number of patient in each snf in 2019
data_snf_patientnum_year = data.groupby(['PID'])[['EpisodeID']].count()
```

```python
data_snf_patientnum_year
```

```python
#load dataset
complete_dataset = pd.read_csv("Data/curr_complete_dataset_delete_columns.csv")
```

```python
complete_dataset.head()
```

```python
# combine the number of patient column into complete_dataset and create a new data frame
result = pd.merge(data_snf_patientnum_year, complete_dataset, how='left', on='PID')
```

```python
result.rename(columns={'EpisodeID':'patient_number_year'}, inplace = True)
```

```python
result.head()
```

```python
# load CCs.csv dataset
cc_number = pd.read_csv("Data/CCs.csv")
```

```python
# add relative columns from CCs.csv into current complete dataset
result2 = pd.merge(result, cc_number, how='left', on='PID')
```

```python
result2.head()
```

```python
result2.to_csv("Data/curr_complete_dataset_delete_columns.csv",index=False)
```
